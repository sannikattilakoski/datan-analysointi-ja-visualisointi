```python
def autot(url_a1, url_a2, url_b1, url_b2):
    import pandas as pd
    import numpy as np
    
    df1 = pd.read_csv(url_a1, sep= ';')
    df2 = pd.read_csv(url_a2, sep= ';')
    df3 = pd.read_csv(url_b1, sep= ';')
    df4 = pd.read_csv(url_b2, sep= ';')
    
    dfa = pd.concat([df1, df2])
    dfb = pd.concat([df3, df4])

    df = pd.merge(dfa, dfb, on = 'ID') 
    df['Ft'].replace(['DIESEL', 'PETROL'], ['Diesel', 'Petrol'], inplace=True)
    
    df= df[['Enedc (g/km)', 'MS', 'Ft']]
    df.rename(columns = {'Enedc (g/km)':'Avg emissions (g/km)', 'MS':'Member State', 'Ft': 'Fuel type'}, inplace=True)
    df = df.groupby(['Member State', 'Fuel type'])['Avg emissions (g/km)'].mean()
    df = df.unstack()
    df = df[['Diesel','Petrol']]
    df.columns = pd.MultiIndex.from_tuples([('Avg emissions (g/km)', 'Diesel'), ('Avg emissions (g/km)', 'Petrol')], names=['','Fuel type'])
    df['Avg emissions (g/km)','D-P'] = df['Avg emissions (g/km)', 'Diesel'] - df ['Avg emissions (g/km)','Petrol']
    df = df.sort_values(('Avg emissions (g/km)','D-P'), ascending = False)
    df = df.fillna(0)
    tulos = df.round().astype(int).head()
    
    print(tulos)

    
### alla olevat rivit taas vain omaan testaukseen, ei palautukseen
url_T5_a1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv'
url_T5_a2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv'
url_T5_b1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB1.csv'
url_T5_b2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB2.csv'

url_T5_a3='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA3.csv'
url_T5_a4='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA4.csv'
url_T5_b3='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB3.csv'
url_T5_b4='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB4.csv'

autot(url_T5_a1, url_T5_a2, url_T5_b1, url_T5_b2)
#autot(url_T5_a3, url_T5_a4, url_T5_b3, url_T5_b4)

# PS tehtävässä pitää yllättäen esiintyä pandas
```
